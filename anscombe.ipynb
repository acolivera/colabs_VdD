```python
#ejemplo de la importancia de la visualización
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Anscombe's quartet dataset from Seaborn's built-in data
df = sns.load_dataset("anscombe")

# Print the first few rows of the DataFrame
print(df.head())

# Create a visualization that shows the regression line within each dataset
sns.lmplot(
    data=df, x="x", y="y", col="dataset", hue="dataset",
    col_wrap=2, palette="muted", ci=None, height=4,
    scatter_kws={"s": 50, "alpha": 1}
);

# Display the plots
plt.show()
```
